```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv("../data/sample_tickets.csv")
df.head()
```

```python
df[['text', 'category']] = df[['text', 'category']].apply(lambda col: col.str.lower())
```

```python
df.head()
```

```python
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    try:
        spacy.cli.download("en_core_web_sm")
        nlp = spacy.load("en_core_web_sm")
    except Exception:
        # fallback to a blank English pipeline if download fails
        nlp = spacy.blank("en")
df['text'] = df['text'].apply(lambda text:[tokens.lemma_ for tokens in nlp(text) if not tokens.is_stop and not tokens.is_punct and not tokens.is_alpha])
df['text'].head()
```
